## Impor library yang diperlukan

```python
import tensorflow as tf                # Library untuk mengakses dataset MNIST dan fungsi ML
from sklearn.model_selection import train_test_split  # Fungsi untuk membagi dataset secara acak
import numpy as np                     # Library untuk operasi numerik
```

## Memuat dataset MNIST

```python
(x_train_full, y_train_full), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# Memuat 60.000 data pelatihan penuh dan 10.000 data pengujian
```

```python
# Normalisasi data
x_train_full = x_train_full / 255.0    # Mengubah nilai piksel dari 0-255 menjadi 0-1 untuk pelatihan
x_test = x_test / 255.0                # Normalisasi data pengujian dengan cara yang sama
```

```python
# Membagi dataset pelatihan penuh menjadi training dan validation
x_train, x_val, y_train, y_val = train_test_split(x_train_full, y_train_full, test_size=0.2, random_state=42)
# 80% untuk training (48.000), 20% untuk validation (12.000), random_state untuk konsistensi
```

```python
# Menampilkan ukuran dataset
print(f'Jumlah data training: {len(x_train)}')    # Menampilkan jumlah sampel di dataset training
print(f'Jumlah data validation: {len(x_val)}')    # Menampilkan jumlah sampel di dataset validation
print(f'Jumlah data testing: {len(x_test)}')      # Menampilkan jumlah sampel di dataset testing
```
